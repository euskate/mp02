```python
# 파일 이름.확장자 명 
path = 'maindata_moo.csv'
# 업로드 변수는 df로 통일 
df = pd.read_csv('./data/'+ path, encoding='utf-8')
```

```python
# 월별 추출 
def get_month(year,month):
    # 구별
    if month <10 : 
        month = str(month) 
        month = '0'+ month
    elif month >= 10:
        month = str(month)
    # 포문        
    li_year = [ i[0] for i in df.날짜.str.split('-') ]
    li_month = [ i[1] for i in df.날짜.str.split('-') ]

    # 
    idx = list()
    for i in range(len(li_year)):
        if li_year[i] == str(year):
            if li_month[i] == month:
                idx.append(i)
    
    each_ye_mon = df.iloc[idx]
    
    return each_ye_mon

# 연도 추출 
def get_year(year):
    # 포문 
    li_year = [ i[0] for i in df.날짜.str.split('-') ]
#     li_month = [ i[1] for i in main_data.날짜.str.split('-') ]

    li = df.shape[0]
    idx = list()

    for i in range(li):
        if li_year[i] == str(year):
            idx.append(i)
    #         if li_month[i] == "01":
    #             idx.append(i)

    each_year = df.iloc[idx]
    
    return each_year
```

```python
# 막대를 사용하여 각 범주 형 빈의 관측치 수를 표시
# 중첩된 변수의 갯수를 비교가능 
sns.countplot(x="avg_moo", data=df)
plt.title("평균금액")
plt.show()
```

```python
ax = sns.jointplot(x="avg_moo", y="avgY_moo", data=df, kind="kde")
plt.suptitle("Joint Plot 과 Kernel Density Plot", y=1.02)
plt.show()

ax.savefig('./output/Joint Plot_Kernel Density Plot.png')
```

```python
np.random.seed(0)
sns.stripplot(x="날짜", y="금액", data=df, jitter=True)
plt.title("요일 별 전체 팁의 Strip Plot")
plt.show()
```

# 상관계수 

```python
pair_grid = sns.PairGrid(df) 
pair_grid = pair_grid.map_upper(sns.regplot) 
pair_grid = pair_grid.map_lower(sns.kdeplot) 
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) 

plt.show()

pair_grid.savefig('./output/페어그래프.png')
```
