### apply method로 열 추가
- apply 파라미터에 함수(람다함수)가 오면 된다.

~~~

def answer(vote):
    if vote >= 200000:
        return 1
    return 0
# votes의 모든 개체에 answer함수 적용 후 answer 칼럼 생성, 적재
df["answer"] = df['votes'].apply(answer)
~~~
